```python
#4. Longest trip duration.

q = """
        SELECT *,CAST(STRFTIME('%s',Lpep_dropoff_datetime) as integer) - CAST(STRFTIME('%s',lpep_pickup_datetime) as integer) as second
        FROM nyc_data_sep
        order by second DESC
        limit 1;


    """

sqldf(q)
```

```python
#   5. Mean tip by hour

q = """
        SELECT STRFTIME('%H',Lpep_dropoff_datetime) as hour ,AVG (Tip_amount) as AVG_TIP
        FROM nyc_data_sep
        GROUP by hour
        order by 2 desc

    """

sqldf(q)
```

```python
#7. Average total trip by day of week (Fortunately, we have day of week information. Otherwise, 
# we need to create a new date column without hours from date column. Then, we need to create "day of week" column, i.e Monday, Tuesday .. or 1, 2 .., 
# from that new date column. Total trip count should be found for each day, lastly average total trip should be calculated for each day).

q = """
        
        
        
        select lpep_pickup_day_of_week,avg(trip) from (SELECT lpep_pickup_day_of_week,STRFTIME('%Y%m%d',lpep_pickup_datetime) as date, count(*) as trip
        FROM nyc_data_sep group by date)
        group by lpep_pickup_day_of_week
        


    """

sqldf(q)



```

```python
q = """
        WITH dayly_table AS  (SELECT lpep_pickup_day_of_week,STRFTIME('%Y%m%d',lpep_pickup_datetime) as date, count(*) as trip
            FROM nyc_data_sep group by date)


        select lpep_pickup_day_of_week,avg(trip) from dayly_table
        group by lpep_pickup_day_of_week

    """

sqldf(q)
```

```python
#8. Count of trips by hour (Luckily, we have hour column. Otherwise, a new hour column should be created from date column, then count trips by hour).


q = """
        SELECT STRFTIME('%H',lpep_pickup_datetime) as date, count(*) as trip
            FROM nyc_data_sep group by date



    """

sqldf(q)
```
