15. Find the trips of which trip distance is greater than 15 miles (included) or less than 0.1 mile (included). It is possible to write this with only one where statement.**However**, this time write two queries and "union" them. The purpose of this question is to use union function.
You can consider this question as **finding** outliers in a quick and dirty way, which you would do in your professional life too often.

```python
q = """
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance >=  15
      
  
        
    """
sqldf(q)
```

```python
q = """
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance <= 0.1
      
  
        
    """
sqldf(q)
```

```python
q = """
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance <= 0.1

        UNION
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance >= 15
      
  
        
    """
sqldf(q)
```

```python
q = """
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance <= 0.1

        UNION all
        SELECT * 
        FROM nyc_data_sep
        where Trip_distance >= 15
      
  
        
    """
sqldf(q)
```

16. We would like to see the distribution (not like histogram) of Total_amount. Could you create buckets, or price range, for Total_amount and find how many trips there are in each buckets? Each range would be 5, until 35, i.e. 0-5, 5-10, 10-15 … 30-35, +35. The expected output would be as follows:

```python
q = """
        SELECT * ,
            CASE when Total_amount < 5 then '0-5'
                 when Total_amount >= 5  and Total_amount < 10 then '5-10'
                 when Total_amount >= 10 and Total_amount < 15 then '10-15'
                 when Total_amount >= 15 and Total_amount < 20 then '15-20'
                 when Total_amount >= 20 and Total_amount < 25 then '20-25'
                 when Total_amount >= 25 and Total_amount < 30 then '25-30'
                 when Total_amount >= 30 and Total_amount < 35 then '30-35'
                 when Total_amount >= 35 then '+35'
            END AS PAYMENT_RANGE_in_dollars
        
            
        FROM nyc_data_sep
       
      
  
        
    """
sqldf(q)
```

```python
q = """
        SELECT COUNT(*) as Trip_Count ,
        
            CASE when Total_amount < 5 then '0-5'
                 when Total_amount >= 5  and Total_amount < 10 then '5-10'
                 when Total_amount >= 10 and Total_amount < 15 then '10-15'
                 when Total_amount >= 15 and Total_amount < 20 then '15-20'
                 when Total_amount >= 20 and Total_amount < 25 then '20-25'
                 when Total_amount >= 25 and Total_amount < 30 then '25-30'
                 when Total_amount >= 30 and Total_amount < 35 then '30-35'
                 when Total_amount >= 35 then '35+'
            END AS PAYMENT_RANGE_in_dollars    
        FROM nyc_data_sep

        GROUP BY PAYMENT_RANGE_in_dollars
        ORDER BY PAYMENT_RANGE_in_dollars
       
      
  
        
    """
sqldf(q)
```

17. We also would like to analyze the performance of each driver’s earning. Could you add driver_id to payment distribution table?  The expected output would be as follows:

```python
q = """
        SELECT driver_id, COUNT(*) as Trip_Count ,
        
            CASE when Total_amount < 5 then '0-5'
                 when Total_amount >= 5  and Total_amount < 10 then '5-10'
                 when Total_amount >= 10 and Total_amount < 15 then '10-15'
                 when Total_amount >= 15 and Total_amount < 20 then '15-20'
                 when Total_amount >= 20 and Total_amount < 25 then '20-25'
                 when Total_amount >= 25 and Total_amount < 30 then '25-30'
                 when Total_amount >= 30 and Total_amount < 35 then '30-35'
                 when Total_amount >= 35 then '35+'
            END AS PAYMENT_RANGE_in_dollars    
        FROM nyc_data_sep

        GROUP BY PAYMENT_RANGE_in_dollars,driver_id
        ORDER BY driver_id
       
      
  
        
    """
sqldf(q)
```
