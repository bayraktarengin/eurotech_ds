```python
#   1. Most expensive trip (total amount).
# Subquery 

q = """
        SELECT * 
        FROM nyc_data_sep
        where Total_amount=(SELECT  MAX(Total_amount)
        FROM nyc_data_sep)
    """

sqldf(q)
```

```python
#   1. Most expensive trip (total amount).
# Subquery 

q = """

        select Trip_distance from
        (SELECT * 
        FROM nyc_data_sep
        where Total_amount=(SELECT  MAX(Total_amount)
        FROM nyc_data_sep)) 
    """

sqldf(q)
```

```python
#   2. Most expensive trip per mile (total amount/mile).


q = """
        SELECT MAX(Total_amount/Trip_distance) as Max_exp
        FROM nyc_data_sep

    """

sqldf(q)

```

```python
#   2. Most expensive trip per mile (total amount/mile).
# Subquery


q = """
        SELECT *
        FROM (select MAX(Total_amount/Trip_distance) from nyc_data_sep )

    """

sqldf(q)

```

```python
#   3. Most generous trip (highest tip).

q = """
        SELECT *
        FROM nyc_data_sep where Tip_amount= (select MAX(Tip_amount) from nyc_data_sep )

    """

sqldf(q)


```

```python
#4. Longest trip duration.

q = """
        SELECT *,CAST(STRFTIME('%s',Lpep_dropoff_datetime) as integer) - CAST(STRFTIME('%s',lpep_pickup_datetime) as integer) as second
        FROM nyc_data_sep
        order by second DESC
        limit 1;


    """

sqldf(q)
```

```python
#   5. Mean tip by hour

q = """
        SELECT STRFTIME('%H',Lpep_dropoff_datetime) as hour ,AVG (Tip_amount) as AVG_TIP
        FROM nyc_data_sep
        GROUP by hour
        order by 2 desc

    """

sqldf(q)
```
