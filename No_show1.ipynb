```python
import pandas as pd
import os
file_path=os.getcwd()+'\\data\\no_show.csv' #os.path.join(os.getcwd(),'data','no_show.csv')
df=pd.read_csv(file_path)
df.head()
```

Verinin Seviyesi
low level bir veri verildiğinde, bu veriyi belli seviyelere çekmek gerekir. 

```python
print(df["AppointmentID"].nunique(), df["PatientId"].nunique())
```

```python
df_filt=df.groupby('PatientId',as_index=False).agg({"AppointmentID":"count"})
#Her bir satır Patient ID olsun, o satırda kaç tane appointment ID var onu yazalım.


df_filt=df_filt[df_filt["AppointmentID"]>1]

df_filt.rename(columns={"AppointmentID":"Total"},inplace=True)
df_filt.head()
```

```python
df_noshow=df.groupby(["PatientId","No-show"],as_index=False).agg({"AppointmentID":"count"})

df_noshow.head()
```

```python
df_merged=pd.merge(df_filt,df_noshow,on="PatientId")
df_merged.tail()
```

```python
df_merged.sort_values('PatientId',inplace=True)
df_merged.tail(10)
```

```python
#Her hastanın gelme oranını gösteren sütun ekleyelim:

df_merged["no_show_percent"]=df_merged["AppointmentID"]/df_merged["Total"]
df_merged.tail(10)
```
