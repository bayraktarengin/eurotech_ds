Find out customers (payment table) who spend more than $150, but only consider payments after February 20, 2007. 
'2007-02-20'

```python
payment.head()
```

```python
q = """
        SELECT customer_id, sum(amount) as amount
        FROM payment
        where amount > 150  and payment_date >'2007-02-20'
        group by customer_id

    """

sqldf(q)
```

```python
filtered_payment=payment[payment["payment_date"]>'2007-02-20'].groupby("customer_id").agg({"amount":"sum"})
filtered_payment[filtered_payment["amount"]>150]
```

```python
q = """
        SELECT customer_id, sum(amount) as amount
        FROM payment
        where payment_date >'2007-02-20'
        group by customer_id
        having sum(amount) > 150 

    """

sqldf(q)
```

find out the films (film IDs) that have an average rental duration greater than 5 days, but we're only interested in those films that have a rental rate less than or equal to $2.99.

```python
df.head()
```

```python
q = """
        SELECT 	film_id	, avg(duration)
        FROM df
        where  rental_rate <=2.99
        group by film_id	
        having avg(duration)>5 
    """

sqldf(q)
```
