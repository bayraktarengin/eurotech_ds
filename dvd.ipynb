Find out customers (payment table) who spend more than $150, but only consider payments after February 20, 2007.

```python
q = """
        SELECT customer_id, sum(amount) as amount
        FROM payment
        where amount > 150  and payment_date >'2007-02-20'
        group by customer_id

    """

sqldf(q)
```

Question 25:
Find out the most rented film categories:


```python
q = """

SELECT 
        c.name as category_name, 
        count(r.rental_id) as rental_count
        FROM category c
        JOIN  film_category f on c.category_id=f.category_id
        JOIN inventory i on f.film_id=i.film_id
        JOIN rental r on i.inventory_id=r.inventory_id
        GROUP BY  c.name
        ORDER BY rental_count desc
        


    """

sqldf(q)
```

  SELECT
    f.title,
    f.replacement_cost,
    c.category_id,
    AVG(f.replacement_cost) OVER (PARTITION BY c.category_id) AS avg_rep_cost_by_cat
  FROM film f
  LEFT JOIN film_category fc ON f.film_id = fc.film_id
  LEFT JOIN category c ON c.category_id = fc.category_id
)

SELECT
  title
FROM tbl
WHERE replacement_cost > avg_rep_cost_by_cat

```python
q = """


WITH tbl AS (
  SELECT
    f.title,
    f.replacement_cost,
    c.category_id,
    AVG(f.replacement_cost) OVER (PARTITION BY c.category_id) AS avg_rep_cost_by_cat
  FROM film f
  LEFT JOIN film_category fc ON f.film_id = fc.film_id
  LEFT JOIN category c ON c.category_id = fc.category_id
)

SELECT
  title
FROM tbl
WHERE replacement_cost > avg_rep_cost_by_cat

""" 

sqldf(q)
```
